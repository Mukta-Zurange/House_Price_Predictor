```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
df = pd.read_csv("C:/Users/mukta/Downloads/House_Price_dataset.csv")
df
```

```python
print(df.isnull().sum())
```

```python
df.describe()
```

```python
df.shape
```

```python
df = df.fillna(df.median(numeric_only=True))
df['furnishingstatus'] = df['furnishingstatus'].fillna(df['furnishingstatus'].mode()[0])
df
```

```python
print(df.isnull().sum()) 
```

```python
df['furnishingstatus'].unique()
```

```python
df['mainroad'] = df['mainroad'].map({'yes':1, 'no':0})
df['guestroom'] = df['guestroom'].map({'yes':1, 'no':0})
df['basement'] = df['basement'].map({'yes':1, 'no':0})
df['hotwaterheating'] = df['hotwaterheating'].map({'yes':1, 'no':0})
df['airconditioning'] = df['airconditioning'].map({'yes':1, 'no':0})
df['prefarea'] = df['prefarea'].map({'yes':1, 'no':0})
df['furnishingstatus'] = df['furnishingstatus'].map({'furnished':2, 'semi-furnished':1, 'unfurnished':0})
df
```

```python
sns.pairplot(df)
```

```python
x = df[['area']]    
y = df['price']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
y_train_pred = model.predict(x_train)
y_test_pred = model.predict(x_test)
print("Simple Regression Coefficient (Area):", model.coef_)
print("Intercept: ", model.intercept_)
```

```python
print("Training:")
MSE = mean_squared_error(y_train, y_train_pred)
print("MSE:", MSE)
RMSE = np.sqrt(MSE)
print("RMSE:", RMSE)
R2 = r2_score(y_train, y_train_pred)
print("R^2:", R2)
```

```python
print("Testing:")
MSE = mean_squared_error(y_test, y_test_pred)
print("MSE:", MSE)
RMSE = np.sqrt(MSE)
print("RMSE:", MSE)
R2 = r2_score(y_test, y_test_pred)
print("R^2:",R2)
```

```python
plt.scatter(x_test, y_test, color='blue', label="Actual")

plt.plot(x_test, y_test_pred, color='red', label="Predicted")
plt.xlabel("area")
plt.ylabel("price")
plt.legend()
plt.show()
```

```python
print("Enter the details to predict house price:")

area = float(input("Area (numeric): "))
predicted_price = model.predict([[area]])
print("Predicted price depending on area: ", predicted_price)
```

```python
X = df.drop(columns=['price'])
y = df['price']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
model_multi = LinearRegression()
model_multi.fit(X_train, y_train)
y_pred_multi = model_multi.predict(X_test)
```

```python
print("Multiple Regression Results:")
MSE = mean_squared_error(y_test, y_pred_multi)
print("MSE:", MSE)
RMSE = np.sqrt(MSE)
print("RMSE:", RMSE)
R2 = r2_score(y_test, y_pred_multi)
print("R^2:", r2_score(y_test, y_pred_multi))
```

```python
import joblib
joblib.dump(model_multi, 'model.pkl')
print("Model saved successfully!")
```
